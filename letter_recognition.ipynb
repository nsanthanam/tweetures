Steps

1. Train logistic regression model
2. Compute dot product
3. Feed into sigmoid function
4. Compute log odds

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
import numpy as np
```

```python
fname = '/Users/navaneethan/Dropbox/projects/freshdesk/data/uci_data.csv'
uci = pd.read_csv(fname)
```

```python
X, y = uci.drop(labels='label', axis=1), uci['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X=X_train, y=y_train, )
```

```python
log_reg_coef = log_reg.coef_
```

```python
X_test2 = X_test.as_matrix()
```

```python
dot_prod = np.dot(log_reg_coef, X_test2.transpose())
```

```python
probs = 1/(1 + np.exp(-1*dot_prod))
```

```python
log_odds = np.log(probs/(1-probs))
```

```python
log_odds_sorted = sorted(log_odds)
```
